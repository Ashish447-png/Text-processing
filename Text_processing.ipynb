# Tokenization

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize


```

```python
# sample
text_word = "I love programming"
sent_word = "i love programming, and i also love python. It is my passion"

token_word = word_tokenize(text_word)
print(token_word)
token_sent = sent_tokenize(sent_word)
print(token_sent)
```

```python
for i in token_sent:
  x = word_tokenize(i)
  print(x)
```

# Stopword Removal

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
# english language stop word
stop_words = set(stopwords.words('english'))
print(stop_words)
len(stop_words)

# for spanish language stop_word
stop_words = set(stopwords.words('spanish'))
print(stop_words)
len(stop_words)

# german language
stop_words = set(stopwords.words('german'))
print(stop_words)
len(stop_words)

#Russian language
stop_words = set(stopwords.words('russian'))
print(stop_words)
len(stop_words)
```

```python
text = "i haven't thought this through by doing most of these against themselves"
len(text)
```

```python
# english language stop word
stop_words = set(stopwords.words('english'))

token = word_tokenize(text)

filterd_tokens = [word for word in token if word.lower() not in stop_words]
print(filterd_tokens)
```

# **Stemming**

```python
# NLTK using Stemming (Porter Stemmer)

from nltk.stem import PorterStemmer

# sample
words = ['running', 'burnt', 'happily', 'lovely']

# Stemmer
stemmer = PorterStemmer()

# Apply stemming
stemmed_words = [stemmer.stem(w) for w in words]

print(stemmed_words)



# 2nd example
words = ['running', 'run', 'runner']

# Stemmer
stemmer = PorterStemmer()

# Apply stemming
stemmed_words = [stemmer.stem(w) for w in words]

print(stemmed_words)
```

# **N-gramms**

```python
# 2 grams pair
from nltk.util import ngrams
sample_text = "My name is XYZ, i have a 10 sec record in running 100 meters"
tokonize = word_tokenize(sample_text)

bigrams = list(ngrams(tokonize, 2))
print(bigrams)
```

```python
# 3 grams pair
from nltk.util import ngrams
sample_text = "My name is XYZ, i have a 10 sec record in running 100 meters"
tokonize = word_tokenize(sample_text)

trigrams = list(ngrams(tokonize, 3))
print(trigrams)
```
